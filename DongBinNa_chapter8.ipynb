# 피보나치 재귀 

```python
def fibo(n):
    if n==2 or n==1:
        print("test")
        return 1
    else:
        return fibo(n-1)+fibo(n-2)
```

```python
fibo(4) # 1 1 2 3 5 8  
```

```python
import time
```

```python
start = time.time()
print(fibo(20))
finish =time.time()
print(finish-start)
```

# memoization 피보나치 재귀

```python
n=20
d=[0]*(n+1)

def fibo(x):
    if x==1 or x==2:
        print('test')
        return 1
    if d[x]!=0:
        return d[x]
    d[x] = fibo(x-1)+fibo(x-2)
    return d[x]
```

```python
start = time.time()
print(fibo(n))
finish =time.time()
print(finish-start)
```
