```python
# 5 5 3 1 5 -> 3
# 5 5 3 1 2 5 -> 3
a=[]
b=min(A)
a.append(b)
for i in range(len(A)-1):
    if A[i+1] < A[i] and A[i+1] > a[-1] :
        b=A[i+1]
        a.append(b)
    elif A[i]== a[-1]+1:
        a.append(A[i])
    elif a[-1]==max(A):
        for j in range(i,len(A)-1):
            if A[j] > A[j+1] and A[j] < a[-1] :
                a.append(A[j])
        if A[-1] < a[-1] :
            a.append(A[-1])
                
a       
```
